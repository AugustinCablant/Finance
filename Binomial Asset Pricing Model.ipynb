```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def binomial_asset_pricing(initial_value, d, u, proba = 0.5, iter = 10000):
    """ Binomial Asset Pricing 
    0 <d< 1 < u
    dS0 represents a downward movement
    uS0 represents an upward movement
    """
    asset_values = [initial_value]
    for _ in range(iter):
        if np.random.random() > proba:
            new_asset_value = asset_values[-1] * u
        else:
            new_asset_value = asset_values[-1] * d
        asset_values.append(new_asset_value)
    plt.figure(figsize=(8,6))
    plt.scatter(0, initial_value, marker = 'x', label = 'initial value', color = 'red')
    plt.plot(range(iter + 1), asset_values, label = "Asset values", color = 'black')
    plt.title(f"Illustration of the Binomial Asset Pricing model with (d, u, proba) = ({d}, {u}, {proba}) ")
    plt.xlabel("Iterations")
    plt.ylabel("Values")
    plt.show()
    return None
```

```python
initial_value = 150
d = 0.50
u = 1 / d
proba = 1 / 4
binomial_asset_pricing(initial_value, d, u, proba = proba)
```
